```python
import tensorflow as tf
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.model_selection import train_test_split
```

```python
paths = glob.glob('./dataset/*/*.png')
paths = np.random.permutation(paths)
```

```python
paths = np.random.permutation(paths)
```

```python
image = cv2.imread(paths[0], cv2.IMREAD_GRAYSCALE)
plt.imshow(image)
```

```python
data_x = list()
```

```python
for i in range(len(paths)):
    image = cv2.imread(paths[i])
    #image = cv2.imread(paths[i], cv2.IMREAD_GRAYSCALE)
    data_x.append(image)
```

```python
data_x = np.array(data_x)
```

```python
paths[i].split('\\')[1]
```

```python
data_y = [[(0 if paths[i].split('\\')[1] == 'other_rgb' else 1)] for i in range(len(paths))]
data_y = np.array(data_y)
print(data_y)
```

```python
print(data_x.shape, data_y.shape)
```

```python
data_x = data_x.reshape(data_x.shape[0], 100, 100, 3)
#data_y = pd.get_dummies(data_y)
print(data_x.shape, data_y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.4, shuffle=False, random_state=1004)
```

```python
X = tf.keras.layers.Input(shape=[100, 100,3])

H = tf.keras.layers.Conv2D(32, kernel_size=3, strides=1, padding = 'same',activation='relu')(X)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Conv2D(64, kernel_size=3, strides=1,padding = 'same', activation='relu')(H)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Conv2D(128, kernel_size=3, strides=1,padding = 'same', activation='relu')(H)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Flatten()(H)
H = tf.keras.layers.Dense(512, activation='relu')(H)
Y = tf.keras.layers.Dense(1, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train,batch_size=64,epochs=100, validation_data=(X_test, y_test))
```
